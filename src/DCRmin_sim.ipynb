```python
solid_angle = math.pi*math.sin(half_angle)**2; 
d_area = (25e-6)*(10e-6); #square detector

def Bw(T):
    return 2*h*c**2/WL**5/(np.exp(h*c/(WL*kB*T))-1)

def Nph(T):
    return solid_angle*d_area*(WL/(h*c))*Bw(T)
```
